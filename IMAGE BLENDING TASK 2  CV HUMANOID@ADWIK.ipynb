```python
url1 = "/content/drive/MyDrive/Virat kohli image 1/ViratKohliPerth.jpg"
url2 = "/content/drive/MyDrive/Virat kohli image 1/Virat Kohli final.jpeg"

imageA = io.imread(url1) 
imageA_2 = cv.cvtColor(imageA, cv.COLOR_BGR2RGB) 

imageB = io.imread(url2) 
imageB_2 = cv.cvtColor(imageB, cv.COLOR_BGR2RGB) 

#cv2_imshow(imageA)
#print(imageA.shape)

#cv2_imshow(imageB)
#print(imageB.shape)

cv2_imshow(imageA_2)
print(imageA_2.shape)

cv2_imshow(imageB_2)
print(imageB_2.shape)
```
